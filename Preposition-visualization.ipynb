# Sanghamitra Preposition Visualization

## 1. Importing of the data

```python
# importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as ex

import seaborn as sns
```

```python
df = pd.read_csv("sanghamitra-prepositions - Sheet2.csv")
df
```

```python
# set index column as learner/type
df = df.set_index('learner/type')
df
```

```python
# count null values in each row
df.isnull().sum()
```

```python
# count null value in each column
df.isnull().sum(axis=1)
```

```python
# simple
df = df.apply(lambda x: x.fillna(x.mean()), axis=1)
df
```

```python
# create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap="Blues")
plt.show();
# swap axes
# add avaerge column
# switch grib color - red, and swap the color range
```

```python
# Select the row with index 'shignesh'
row_data = df.loc['shignesh']

# Create a bar chart
plt.figure(figsize=(10, 6))
row_data.plot(kind='bar')
plt.title('Percentage of values for the row shignesh')
plt.xlabel('Column Index')
plt.ylabel('Percentage')
plt.ylim(0, 100)
plt.show();
```

```python
# Create a bar chart for each row
for index, row in df.iterrows():
    plt.figure(figsize=(10, 6))
    row.plot(kind='bar')
    plt.title(f'Percentage values for the row {index}')
    plt.xlabel('Column Index')
    plt.ylabel('Percentage')
    plt.ylim(0, 100)
    plt.show();
    
# how to add drop down menu
# make these in grid charts 
# use facet plot
# check bokey library
```

```python
# Calculate the average values for each column
average_values = df.mean()

# Add a new row 'Average' to the DataFrame
df.loc['Average'] = average_values
```

```python
# Create a bar chart for the 'Average' row using matplotlib
plt.figure(figsize=(10, 6))
df.loc['Average'].plot(kind='bar')
plt.title('Average values for each column')
plt.xlabel('Column Index')
plt.ylabel('Average')
plt.ylim(0, 100)
plt.show()
```

```python
# Create a bar chart for the 'Average' row using Plotly Express
fig = ex.bar(x=df.columns, y=df.loc['Average'], labels={'x':'Column Index', 'y':'Average'})
fig.update_layout(title='Performance of the group')
fig.update_yaxes(range=[0, 100])

# Add a horizontal line for the overall average value
fig.add_hline(y=df.loc['Average'].mean(), line_dash="dash", line_color="red")

fig.show()
```

```python
# Calculate the average values for each row
average_values_rows = df.mean(axis=1)

# Create a bar chart for the average values of each row using matplotlib
plt.figure(figsize=(12, 6))
plt.bar(df.index, average_values_rows)
plt.axhline(y=average_values_rows.mean(), color='r', linestyle='--', label='Overall Average')
plt.xlabel('Row Index')
plt.ylabel('Average')
plt.title('Average values for each row')
plt.legend()
plt.grid(True)
plt.show()
```
